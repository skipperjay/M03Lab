```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
df = yf.download("MSFT",
                 start="2020-01-01",
                 end="2020-12-31",
                 auto_adjust=False,
                 progress=False)
df["simple_rtn"] = df["Adj Close"].pct_change()
df = df.dropna()
```

```python
df.dropna()
```

```python
df["Adj Close"].plot(title="MSFT stock in 2020")
```

```python
(
    df[["Adj Close", "simple_rtn"]]
    .plot(subplots=True, sharex=True,
          title="MSFT stock in 2020")
)
```
